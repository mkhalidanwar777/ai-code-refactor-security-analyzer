### 1 star **Right-angled triangle pattern**


```python
num=int(input('enter a number'))
i=1
while(i<=num+1):
    print('*'*i)
    i+=1
```

### 2 **Inverted triangle pattern**


```python
num = int(input("enter a number: "))

i = num
while i >= 1:
    j = 1
    while j <= i:
        print("*", end="")
        j += 1
    print()
    i -= 1
```

### 6 **Descending number pattern**


```python
num=int(input('enter a number '))
i=num
while(i>=1):
    print(f'{i} ',end='')
    i-=1
```

### 7 Sum of First N Numbers

```python
num=int(input('enter a number '))
result=0
i=1
while(i<=num):
    result=result+i
    i+=1
print(f'You enter a number is : {num}')
print(f'sum of N number is : {result}')
```

### 10 Fibonacci Series

```python
num=int(input(' enter a number  for Fibonacci Series'))
a=0
b=1
i=1
while(i<=num):
    print(f'{a} ',end='')
    temp=a+b
    a=b
    b=temp
    i+=1
```

### 11 check it is prime number or not

```python
num = int(input("enter a number for to check it is prime number or not  : "))
count=0
i=1
while(i<=num):
    if(num%i==0):
        count+=1
    i+=1
if count==2:
    print(f'number is Prime number : {num}')
    
else:
    print(f'Number is not prime number : {num}')
```

### 1 star **Right-angled triangle pattern**


```python
num=int(input('enter a number'))
for i in range(1,num+1):
    for j in range(1,i+1):
        print('*',end='')
    print()
```

### 2 **Inverted triangle pattern**


```python
num=int(input('enter a number'))
for i in range(1,num+1):
    print('*'*num)
    num=num-1
```

### 7 Sum of First N Numbers

```python
num=int(input('enter a number '))
result=0
for i in range(1,num+1):
    result=result+i
print(f'sum of N number is : {result}')
```

### 8 Factorial Finder

```python
num=int(input('enter a number for to find the Factorial'))
result=1
for i in range(num,1,-1):
    result=result*i
print(f'the factorial of {num} is : {result}')
```

### 10 Fibonacci Series

```python
num=int(input(' enter a number  for Fibonacci Series'))
a=0
b=1
for i in range(num):
    print(f'{a} ',end='')
    temp=a+b
    a=b
    b=temp
```

### 11 check is number is Prime Number or not

```python
num = int(input("enter a number for to check it is prime number or not  : "))
count=0
for i in range(1,num+1):
    if num%i==0:
        count=count+1
if count==2:
    print(f'number is Prime number : {num}')
    
else:
    print(f'Number is not prime number : {num}')
```
